### Опис датасета
This data set represents the behavioral data of a utility app designed for car-related
transactional services. The app is subscription-based, and users are charged monthly or
annually.

The transaction (event «Transaction») is the primary feature that proves the user is getting value
from the product. As long as the user has them - there is an economic sense for this user to
keep paying for the app.

App operates in various regions of the United States, and the service structure differs
significantly from state to state as the service/transaction availability differs for each state.

Also, users have slight differences, some of them need to conduct transactions more
frequently, and some less often. The best example, in this case, would be Uber - some people
need to drive more regularly, and some use the app only occasionally.

The dataset contains data for the year 2022 and has the next events:
- Transaction Refund - user receives a refund, because he probably didn’t perform the Transaction and was charged by mistake
- Order - target transaction performed, also has multiple event properties included
- Account Setup Skip - users skip registration
- Signup Success - successful registration, includes information about the Signup method - social network, email, etc
- Add Vehicle Success - vehicle registration, part of the onboarding process, also includes the information about the vehicle.
- Add Vehicle Break - user didn’t complete vehicle registration
- Add Payment Method Sucess - user adds the credit card
- Add Payment Method Failed - user experiences failure while adding the payment method
- Email Confirmation Success - user passes through the email veri cation process, now we are user he receives messaging from the product
- Chat conversation Opened - user contacts the Support, that’s usually is not good, something is probably going wrong
- Sign Out - user signs out from the account, maybe he even wants to terminate the subscription, but this is not a complete action to close the account
- Account History Transaction Details - user checks some particular transactions, usually he is either savvy and likes to audit the expenses or he has a suspicion that there might be some billing errors
- Wallet Opened - user checks his balance state
- Subscription Premium - user buys a monthly or annual subscription
- Subscription Premium Cancel - explicit churn event - user has canceled the subscription. That’s the event we want to predict and prevent
- Calculator view - users want to get an idea about the cost of future transaction

Також як ми бачимо кожна подія має свої атрибути. Випишемо унікальні атрибути для кожної з подій та унікальні значення кожного з атрибутів. Збережемо ці дані у словнику до змінної event_to_event_attributes_dict. Будемо використовувати у майбутньому для аналізу.

```python
event_to_event_attributes_dict={}
for event in df["event_name"].unique(): #проходимось по кожній події
    temp_dict = {}
    for attributes in df[df["event_name"] == event]["event_attributes"]: #проходимось по кожному атрибуту
        temp_instance_atttributes = json.loads(attributes)
        
        for attribute in temp_instance_atttributes: #зберігаємо унікальне значення по кожному атрибуту
            if attribute in temp_dict:
                if temp_instance_atttributes[attribute] == "":
                    if "NaN" not in temp_dict[attribute]:
                        temp_dict[attribute] = temp_dict[attribute] + ["NaN"]
                    else:
                        pass
                elif temp_instance_atttributes[attribute] not in temp_dict[attribute]:
                    temp_dict[attribute] = temp_dict[attribute] + [temp_instance_atttributes[attribute]]
            else:
                if temp_instance_atttributes[attribute] == "":
                    temp_dict[attribute] = ["NaN"]
                else:
                    temp_dict[attribute] = [temp_instance_atttributes[attribute]]
    event_to_event_attributes_dict[event]=temp_dict
for event in event_to_event_attributes_dict:
    print(f"Атрибути та унікальні значення для події {event}:")
    for attribute in event_to_event_attributes_dict[event]:
        print(f"{attribute} : {event_to_event_attributes_dict[event][attribute]}", end="\n")
    print("\n\n\n")
```

```python
d = {"userid": df["userid"].unique()}

for i in event_to_event_attributes_dict:
    temp_lst=[]
    for j in d["userid"]:
        temp_lst.append(len(df[(df["userid"]==j) & (df["event_name"]==i)]))
    d[i] = temp_lst
    
temp_lst=[]
for j in d["userid"]:
    temp_lst.append(df[df["userid"]==j]["user_state"].unique()[0])
d["user_state"] = temp_lst
    
df_count_of_events = pd.DataFrame(data=d)
```

DecisionTree

```python
params = {
    "max_depth": [3, 5, 7, 9, 11, 13],
}

model_desicion_tree = DecisionTreeClassifier(
    random_state=SEED,
    class_weight='balanced',
)

model_desicion_tree = GridSearchCV(
    model_desicion_tree, 
    params, 
    cv=3,
    scoring='accuracy',
)

model_desicion_tree.fit(X, y)

print(f'Best parameters {model_desicion_tree.best_params_}')
print(
    f'Mean cross-validation accuracy score of the best_estimator: ' + \
    f'{model_desicion_tree.best_score_:.3f}'
)
cv_scores['desicion_tree'] = model_desicion_tree.best_score_
```

Так як GridSearchCV немає атрибутів intercept_ та coef_ натренуємо на найкращих параметрах окрему модель і подивимось на ваги для кожної з ознак. Можливе некоректне відображення імен івентів треба замість пропусків вставити"_".

```python
import math

df_fi["features"]=["_".join(i.split()) for i in df_fi["features"]]

model_logistic_regression = LogisticRegression(
    random_state=SEED,
    class_weight="balanced",
    solver="liblinear",
    C=0.1,
    penalty="l2")
model_logistic_regression.fit(X_norm, y)
w = model_logistic_regression.coef_
df_fi["LR"] = pow(math.e, w)[0] #Задля переформатування від'ємних значень
df_fi["features"]=["_".join(i.split()) for i in df_fi["features"]]
ax = df_fi.plot.barh(x="features", y="LR")
plt.show()
```
